```python
class LinearRegression_farida:
    def __init__(self, w=0.0, b=0.0, lr=0.01, epochs=1000):
        self.w = w
        self.b = b
        self.lr = lr
        self.epochs = epochs

    def fit(self, X, y):
        n = len(X)

        for _ in range(self.epochs):
            y_pred = self.w * X + self.b

            dw = (-2 / n) * sum(X * (y - y_pred))
            db = (-2 / n) * sum(y - y_pred)

            self.w -= self.lr * dw
            self.b -= self.lr * db

    def predict(self, x):
        return self.w * x + self.b
```
