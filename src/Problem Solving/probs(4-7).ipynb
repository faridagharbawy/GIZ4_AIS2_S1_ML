```python
#problem 4: add 2 nums
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def addTwoNumbers(l1, l2):
    dummy = ListNode(0)
    current = dummy
    carry = 0

    while l1 or l2 or carry:
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0

        total = val1 + val2 + carry
        carry = total // 10

        current.next = ListNode(total % 10)
        current = current.next

        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next

    return dummy.next


#test
def build_linked_list(values):
    dummy = ListNode(0)
    current = dummy
    for v in values:
        current.next = ListNode(v)
        current = current.next
    return dummy.next


def linked_list_to_list(node):
    result = []
    while node:
        result.append(node.val)
        node = node.next
    return result


#website's examples
l1 = build_linked_list([2, 4, 3])
l2 = build_linked_list([5, 6, 4])
result = addTwoNumbers(l1, l2)
print(linked_list_to_list(result))  # [7, 0, 8]

l1 = build_linked_list([0])
l2 = build_linked_list([0])
result = addTwoNumbers(l1, l2)
print(linked_list_to_list(result))  # [0]

l1 = build_linked_list([9, 9, 9, 9, 9, 9, 9])
l2 = build_linked_list([9, 9, 9, 9])
result = addTwoNumbers(l1, l2)
print(linked_list_to_list(result))  # [8, 9, 9, 9, 0, 0, 0, 1]


```

```python
#problem 5 : 
def romanToInt(s):
    roman = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
    }
    
    total = 0
    prev = 0
    
    for char in reversed(s):
        value = roman[char]
        
        if value < prev:
            total -= value
        else:
            total += value
        
        prev = value
    
    print (total)

romanToInt("III")
romanToInt("LVIII")
romanToInt("MCMXCIV")
```

```python
#problem 6: earch insert position
def searchInsert(nums, target):
    left = 0
    right = len(nums) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return left


print(searchInsert([1,3,5,6], 5))  # 2
print(searchInsert([1,3,5,6], 2))  # 1
print(searchInsert([1,3,5,6], 7))  # 4
```

```python
#problem 7: calc
def evaluate_expression(s: str) -> int:
    if not s:
        return 0
    
    stack = []
    num = 0
    sign = '+'  # previous operator, default '+'
    s = s.replace(' ', '')  # remove spaces for simplicity
    n = len(s)
    
    for i, c in enumerate(s):
        if c.isdigit():
            num = num * 10 + int(c)
        if c in '+-*/' or i == n - 1:
            if sign == '+':
                stack.append(num)
            elif sign == '-':
                stack.append(-num)
            elif sign == '*':
                stack[-1] = stack[-1] * num
            elif sign == '/':
                # truncate toward zero
                stack[-1] = int(stack[-1] / num)
            sign = c
            num = 0

    return sum(stack)

#test
test_cases = ["3 + 2 * 2", " 3 / 2 ", " 3 + 5 / 2 "]
for expr in test_cases:
    print(f"{expr} = {evaluate_expression(expr)}")

```
