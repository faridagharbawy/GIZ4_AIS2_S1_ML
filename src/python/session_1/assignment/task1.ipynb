```python
#input Validation
mail = "Amit_ml@gmail.edu"
mailsplitted = mail.split("@")
if mail.count("@") == 1 and "." in mailsplitted[1]:
    print("Valid mail")
else:
    print("Invalid mail")
```

```python
#extract username
mailsplitted = mail.split("@")
print("Extracted username is: " + mailsplitted[0])
```

```python
#extract domain
print("Extracted domain is: " + mailsplitted[1])
```

```python
#domain ending
if mail.endswith(".com"):
    print("Commercial Domain")
elif mail.endswith(".edu"):
    print("Educational Domain")
else:
    print("Other Domain")
```
