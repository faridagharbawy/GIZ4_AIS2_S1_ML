```python
#step1> decoding (###!!@mocleW EPGTQ!!!6789)
msg = "###!!@emocleW EPGTQ!!!6789"
extracted = msg[6:19]
print(extracted)
```

```python
#step 2
word1 = extracted.split()[0]
reversed_word = ""
for char in word1:
    reversed_word = char + reversed_word
print(reversed_word)
```

```python
# step 3
word2 = extracted.split()[1]        # "EPGTQ"
if word2[0] == "E":
    word2_decoded = word2[1:]      # "PGTQ"
else:
    word2_decoded = word2
print(word2_decoded)
```

```python
#final
decoded_message = reversed_word + " " + word2_decoded

print(decoded_message)
```
