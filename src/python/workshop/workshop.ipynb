```python
#calculate average student grades
def calc_avg():
    students_num = int(input("Enter number of students"))
    students_grades = []
    
    for i in range (students_num):
        grade = float(input("Enter your grade: "))
        students_grades.append(grade)
    
    if students_num == 0:
        print("No students")
    else:
        print(f'Average of grades is {sum(students_grades) / students_num}')
        
calc_avg()
```
