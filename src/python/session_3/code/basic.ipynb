```python
def factorial(num):
    if num == 0:
        return 1
    if num == 1:
        return 1
    return num * factorial(num-1)

factorial(5)
```

```python
#trying factorial
def facttry(numtry):
    if numtry == 0 or numtry == 1:
        return 1
    return numtry * facttry(numtry - 1)
facttry(5) 
```

```python
def isprime(num):
    for i in range(2,num):
        if num % i == 0:
            return False
    return True
isprime (20)
```

```python
def primetry(primetry):
    for i in range (2, primetry):
        if primetry % i == 0:
            return "Not a prime number"
        return "Prime Number"
primetry(3)
```

```python
def dividors(num1, num2):
    """
    takes 2 numbers, then gets the min among them
    checks if the 2 numbers have remainders or not
    if they are divisible, values are appended inside the empty list
    """
    
    mn = min(num1,num2)
    common_div = []
    for i in range(1, mn+1):
        if num1 % i == 0 and num2 % i == 0:
            common_div.append(i)
    return common_div
dividors(2,4)
```
