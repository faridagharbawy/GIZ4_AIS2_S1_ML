```python
def reverse_vocab(s):
    st_1 = s.split(" ")
    new_st = st_1[::-1]
    new_str = ' '.join(new_st)
    return new_str
reverse_vocab("hi dida")
```

```python
#conv to function
def count_words(words):
    unique_words = set(words)
    counts = {word: 0 for word in unique_words}

    for i in unique_words:
        for j in words:
            if i == j:
                counts[i] += 1   

    return counts

count_words(["Welcome", "Ali", "Hi", "Ali", "No", "Hi", "No", "Ali", "No", "Ali"])
```

```python
#chatbot
import random
responses = {
    "hello" : ["Hello", "Hi" , "Greetings" ],
    "how are you" : ["I am doing well, Thank You", "I am fine, How about you?"],
    "goodbye" : ["Goodbye", "See you later", "Farewell"],
    "default" : ["I am sorry, I did not understand.", "Could you please rephrase what you've said?"]
}

def get_response(user_input):
    for key in responses:
        if key in user_input:
            return random.choice(responses[key])
    return random.choice(responses["default"])

def chatbot():
    print("chatbot: Hello!, How can I help you?")
    
    while True:
        user_input = input("User:   ").lower()
        response = get_response(user_input)
        print("chatbot: ", response)
        
        if user_input == "goodbye":
            break
        
chatbot()
```
