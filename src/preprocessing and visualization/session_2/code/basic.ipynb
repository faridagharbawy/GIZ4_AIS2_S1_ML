```python
#saving to file using (savez) w between brackets esm el file
mat7 = np.ones((3,4))
mat8 = np.zeros((5,3))
np.savez('np_file', ones_matrix = mat7, zeros_matrix = mat8)

with np.load('np_file.npz') as file:
    mat7 = file['ones_matrix']
    mat8 = file['zeros_matrix']
    
print(mat7)
print(mat8)
```

```python
def array_reshape(arr, rows = -1, cols = -1):
    try:
        arr = np.array(arr)
        return arr.reshape(rows, cols)
    except (TypeError, ValueError) as e:
        raise ValueError(f"Invalid reshape operation: {e}")
    
vec = [1,2,3,4,5,6]
print(array_reshape(vec,2,3))
```

```python
#nlef ala kol el elements nshof law greater than 1 > true or false
v = np.array([1,2,3,4,5,0])
print(np.all(v >= 1))
```

```python
#bshof how diff are they, atol bthaded close b ad eh
v1 = np.array([1,2,3,4,5])
v2 = np.array([2,3,0,5,3])
print(np.isclose(v1, v2, atol=1))
print(np.allclose(v1, v2, atol=1))
```

```python
v3 = np.array([2,4,3,4,10])
condition = v3>= 4
print(condition)
```

```python
v4 = np.array([2,4,3,4,10])
condition = v4>= 4
conditional_access = v4[condition]
print(conditional_access)
```

```python
v5 = np.array([2,4,3,4,10])
condition = v5>= 4
new_arr = np.where(condition, -1,v5)
print(new_arr)
```
